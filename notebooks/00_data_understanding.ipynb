```python
import pandas as pd
import os

# Check if files exist before reading
file_paths = {
    'train': r"C:\Users\mahip\Desktop\Retail-Demand-Forecasting-Analytics\data\raw\train(1).csv",
    'store': r"C:\Users\mahip\Desktop\Retail-Demand-Forecasting-Analytics\data\raw\store.csv",
    'test': r"C:\Users\mahip\Desktop\Retail-Demand-Forecasting-Analytics\data\raw\test.csv"
}

# Check which files exist
for name, path in file_paths.items():
    if os.path.exists(path):
        print(f"✓ {name} file found")
    else:
        print(f"✗ {name} file NOT found at: {path}")

# Try to read the files with error handling
try:
    train = pd.read_csv(file_paths['train'])
    print("Train data loaded successfully")
except FileNotFoundError:
    print("Train file not found. Please check the file path and name.")
    # Alternative: try without (1) in filename
    try:
        train = pd.read_csv(r"C:\Users\mahip\Desktop\Retail-Demand-Forecasting-Analytics\data\raw\train.csv")
        print("Train data loaded from 'train.csv'")
    except FileNotFoundError:
        print("Neither 'train(1).csv' nor 'train.csv' found")

try:
    store = pd.read_csv(file_paths['store'])
    print("Store data loaded successfully")
except FileNotFoundError:
    print("Store file not found. Please check the file path.")

try:
    test = pd.read_csv(file_paths['test'])
    print("Test data loaded successfully")
except FileNotFoundError:
    print("Test file not found. Please check the file path.")
```
